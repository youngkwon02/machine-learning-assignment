# K-means clustering

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib import cm
```

## load data

```python
fname_data = 'assignment_11_data.csv'

feature = np.genfromtxt(fname_data, delimiter=',')

x = feature[:,0]
y = feature[:,1]

number_data     = np.size(feature, 0)
number_feature  = np.size(feature, 1)

print('number of data : {}'.format(number_data))
print('number of feature : {}'.format(number_feature))
```

## plot the input data

```python
plt.figure(figsize=(8,8))   
plt.title('input data')

plt.scatter(x, y, color='blue')

plt.tight_layout()
plt.show()
```

## compute distance
- feature : $n \times m$, center : $1 \times m$, distance : $n \times 1$
- $n$ : number of data, $m$ : number of features

```python
def compute_distance(feature, center):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    distance =(feature - center) ** 2
    distance = distance[:, 0] + distance[:, 1]

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return distance
```

## compute centroid
- feature : $n \times m$, label_feature : $n \times 1$, value_label : $1 \times 1$, centroid : $1 \times m$
- $n$ : number of data, $m$ : number of features

```python
def compute_centroid(feature, label_feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    k_clus_feat = feature[np.where(label_feature == label)]
    
    if len(k_clus_feat) == 0:
        return [-1, -1]
    
    centroid = k_clus_feat.sum(axis=0) / len(k_clus_feat)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return centroid
```

## compute label
- distance : $n \times k$, label_feature : $n \times 1$
- $n$ : number of data, $k$ : number of clusters

```python
def compute_label(distance):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    label_feature = np.argmin(distance, axis=1)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return label_feature
```

## the number of clusters $K = 2$

```python
number_cluster          = 2
number_iteration        = 900     # you can modify this value
loss_iteration_02       = np.zeros(number_iteration)
centroid_iteration_02   = np.zeros((number_iteration, number_cluster, number_feature))
label_feature_02        = np.random.randint(0, number_cluster, size=(number_data))
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks
#

def compute_loss(feature, label_feature, label_cluster, centroid):
  loss = 0
  for i in range(len(label_cluster)):
    j = np.where(label_feature == i)
    loss += np.sum(compute_distance(feature[j], centroid[i]))
  return loss / label_feature.shape[0]


dist_iteration_02 = np.zeros(shape=(number_data, number_cluster))
cent = np.zeros((number_cluster, number_feature))

for i in range(number_iteration):
  for c in range(number_cluster):
    cent[c] = compute_centroid(feature, label_feature_02, c)
    dist_iteration_02[:, c] = compute_distance(feature, cent[c])
    
  label_feature_02 = compute_label(dist_iteration_02)
  loss_iteration_02[i] = compute_loss(feature, label_feature_02, np.arange(0, number_cluster), cent)
  centroid_iteration_02[i] = cent

#
# ++++++++++++++++++++++++++++++++++++++++++++++++++
```
